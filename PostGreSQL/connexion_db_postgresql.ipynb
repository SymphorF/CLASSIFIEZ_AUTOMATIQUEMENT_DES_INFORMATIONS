```python
# Traitement de la colonne 'augementation_salaire_precedente'

df_clean['augmentation_salaire_precedent_pct'] = df_clean['augementation_salaire_precedente'].str.split(' ', n=1, expand=True)[0]

df_clean['augmentation_salaire_precedent_pct'] = pd.to_numeric(df_clean['augmentation_salaire_precedent_pct'])

df_clean.drop(columns='augementation_salaire_precedente', inplace=True)
```

### 4 - Encodage des données

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, OrdinalEncoder

# 1. Encodage par LabelEncoder 
le = LabelEncoder() 
df_clean['a_quitte_l_entreprise_encoded'] = le.fit_transform(df_clean['a_quitte_l_entreprise'])
df_clean['ayant_enfants'] = le.fit_transform(df_clean['ayant_enfants'])

# 2. Encodage par OrdinalEncoder
oe = OrdinalEncoder()
def appliquer_ordinal_encoding(df):
    # Définir l’ordre des catégories pour chaque colonne à encoder
    ordre_dict = {'frequence_deplacement': ['Aucun', 'Occasionnel', 'Frequent']}
    for col, order in ordre_dict.items():
        if col in df.columns:
            oe = OrdinalEncoder(categories=[order])
            encoded_values = oe.fit_transform(df[[col]])
            df[f"{col}_encoded"] = encoded_values.astype(int)
        else:
            print(f"⚠️ Colonne '{col}' non trouvée dans le DataFrame.")                                                                         
    return df
df_clean = appliquer_ordinal_encoding(df_clean)


# 3. Encodage par OneHotEncoder
col_to_onehot = ['genre', 
                 'statut_marital', 
                 'departement', 
                 'poste',
                 'domaine_etude', 
                 'heure_supplementaires']
def appliquer_onehot_encoding(df, colonnes):
    # Sélectionner uniquement les colonnes existantes dans le DataFrame
    colonnes_existantes = [col for col in colonnes if col in df.columns]

    if colonnes_existantes:
        ohe = OneHotEncoder(sparse_output=False, drop=None)
        encodage = ohe.fit_transform(df[colonnes_existantes])

        # Créer les noms de colonnes encodées
        noms_colonnes = ohe.get_feature_names_out(colonnes_existantes)

        # Créer un DataFrame avec les colonnes encodées
        df_encoded = pd.DataFrame(encodage, columns=noms_colonnes, index=df.index)

        # Concaténer avec le DataFrame d’origine
        df = pd.concat([df, df_encoded], axis=1)
    else:
        print("⚠️ Aucune des colonnes spécifiées n’est présente dans le DataFrame.")
    return df

df_clean = appliquer_onehot_encoding(df_clean, col_to_onehot)
df_clean.drop(columns=['genre', 
                       'statut_marital', 
                       'departement', 
                       'poste',
                       'domaine_etude', 
                       'heure_supplementaires',
                       'a_quitte_l_entreprise',
                       'ayant_enfants',
                       'frequence_deplacement'], 
                       inplace=True)
```

### 5 - Séparation des données d'entrée de la target X/y

```python
X = df_clean.drop(columns=['a_quitte_l_entreprise_encoded'])
y = df_clean['a_quitte_l_entreprise_encoded']

X.to_sql('model_inputs', engine, if_exists='replace', index=False) # Enregistrement de X dans la base de données
y.to_frame(name="a_quitte_l_entreprise").to_sql('model_target', engine, if_exists='replace', index=False)
```

### 6 - Modélisation

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier 
from sklearn.model_selection import train_test_split

# I. Random Forest Classifier

#  Séparation en train/test
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y,test_size=0.2, random_state=42)

#  Modèle avec max_depth
model_rfc = RandomForestClassifier(
    n_estimators=100,
    max_depth=10,  # 
    random_state=42,
    class_weight='balanced'
)
model_rfc.fit(X_train, y_train)


# II. Regression Logistique

# Modèle
logreg = LogisticRegression(max_iter=5000)
logreg.fit(X_train, y_train)
```

### 7- Prédictions 

```python
prediction_rfc = model_rfc.predict(X_test) # Predictions avec Random Forest

prediction_rl = logreg.predict(X_test) # Prédictions avec Regression Logistique
```

### 8 - Enregistrement dans la base de données

```python

# Construction d’un DataFrame résultats rfc
df_pred = pd.DataFrame({
    "id_employee": X_test['id_employee'],
    "prediction": prediction_rfc})
df_pred.to_sql("model_outputs_rfc", engine, if_exists="replace", index=False)


# Construction d’un DataFrame résultats regression logistique
df_pred_2 = pd.DataFrame({
    "id_employee": X_test['id_employee'],
    "prediction": prediction_rl})
df_pred_2.to_sql("model_outputs_logreg", engine, if_exists="replace", index=False)
```
